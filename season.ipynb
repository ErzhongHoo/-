```python
import pandas as pd

cation_resin_1 = pd.read_csv('authentic_cs_1.csv')
# cation_resin_1['time_diff'] = pd.to_datetime(cation_resin_1['time']).diff()
# cation_resin_1['time_diff'] = cation_resin_1['time_diff'].dt.total_seconds() / 60 / 60

cation_resin_1['date'] = pd.to_datetime(cation_resin_1['date'], format='%Y-%m-%d')

cation_resin_1 = cation_resin_1.set_index('date')

# cation_resin_1_s = cation_resin_1.resample('Y').mean()
cation_resin_1_s = (cation_resin_1.resample('Q').mean()).round(2)
# cation_resin_1_s = cation_resin_1.resample('M').mean()
print(cation_resin_1_s)
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd

plt.figure(
    figsize = (16, 8),
    dpi = 100
)

ydata = cation_resin_1_s['time_diff']
plt.plot(ydata, color='red')

plt.xticks(rotation=45)

plt.xlabel("date")
plt.ylabel("time_diff")
plt.title("1# cation resin")

plt.savefig('cation_resin_1.jpg')

plt.show()
```

### 三合一季节图

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

cation_resin_1 = pd.read_csv('authentic_cs_1.csv')
cation_resin_1['date'] = pd.to_datetime(cation_resin_1['date'], format='%Y-%m-%d')
cation_resin_1 = cation_resin_1.set_index('date')

# cation_resin_1_s = cation_resin_1.resample('Y').mean()
# cation_resin_1_s = (cation_resin_1.resample('Q').mean()).round(2)
cation_resin_1_s = cation_resin_1.resample('M').mean()

cation_resin_2 = pd.read_csv('authentic_cs_2.csv')
cation_resin_2['date'] = pd.to_datetime(cation_resin_2['date'], format='%Y-%m-%d')
cation_resin_2 = cation_resin_2.set_index('date')

# cation_resin_2_s = cation_resin_2.resample('Y').mean()
# cation_resin_2_s = (cation_resin_2.resample('Q').mean()).round(2)
cation_resin_2_s = cation_resin_2.resample('M').mean()

cation_resin_3 = pd.read_csv('authentic_cs_3.csv')
cation_resin_3['date'] = pd.to_datetime(cation_resin_3['date'], format='%Y-%m-%d')
cation_resin_3 = cation_resin_3.set_index('date')

# cation_resin_3_s = cation_resin_3.resample('Y').mean()
# cation_resin_3_s = (cation_resin_3.resample('Q').mean()).round(2)
cation_resin_3_s = cation_resin_3.resample('M').mean()



print(cation_resin_1_s.describe())
print(cation_resin_2_s.describe())
print(cation_resin_3_s.describe())
print((cation_resin_1_s.max()+cation_resin_2_s.max()+cation_resin_3_s.max())/3)


plt.figure(
    figsize = (16, 8),
    dpi = 100
)

y1data = cation_resin_1_s['time_diff']
y2data = cation_resin_2_s['time_diff']
y3data = cation_resin_3_s['time_diff']
plt.plot(y1data, color='red', label='1#')
plt.plot(y2data, color='blue', label='2#')
plt.plot(y3data, color='green', label='3#')

plt.xticks(rotation=45)

plt.xlabel("date")
plt.ylabel("time_diff")
plt.title("cation resin")
plt.legend(loc='lower right') # 标签位置

plt.savefig('cation_resin_monthly.jpg')

plt.show()
```
