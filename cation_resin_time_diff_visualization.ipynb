```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd

data = pd.read_csv('authentic_cs_1.csv')


data['date'] = pd.to_datetime(data['date'], format='%Y-%m-%d')
# data['time_diff'] = pd.to_datetime(data['time']).diff()
# data['time_diff'] = data['time_diff'].dt.total_seconds() / 60 / 60

data.drop([data['time_diff'].idxmax()], axis=0, inplace=True)


plt.figure(
    figsize = (16, 8),
    dpi = 100
)

xdata = data['date']
ydata = data['time_diff']
plt.plot(xdata, ydata, color='red')

plt.xticks(rotation=45)

plt.xlabel("date")
plt.ylabel("time_diff")
plt.title("regeneration info")

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd

data = pd.read_csv('cation_resin_2.csv')

data['date'] = pd.to_datetime(data['date'], format='%Y-%m-%d')
data['time_diff'] = pd.to_datetime(data['time']).diff()
data['time_diff'] = data['time_diff'].dt.total_seconds() / 60 / 60

plt.figure(
    figsize = (16, 8),
    dpi = 100
)

xdata = data['date']
ydata = data['time_diff']
plt.plot(xdata, ydata, color='blue')

plt.xticks(rotation=45)

plt.xlabel("date")
plt.ylabel("time_diff")
plt.title("regeneration info")

plt.show()
```
