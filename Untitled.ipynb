```python
import pandas as pd
wp_data = pd.read_csv("WatarProductionRawInfo.csv")

wp_data['capacity'].loc[wp_data['capacity'] >= 10000] = (wp_data['capacity'].mean()).round(2)
wp_data['capacity'].loc[wp_data['capacity'] < 0] = (wp_data['capacity'].mean()).round(2)
wp_data['capacity'].fillna((wp_data['capacity'].mean()).round(2), inplace=True)

wp_data.to_csv('authentic_capacity.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

wp_capacity = pd.read_csv('authentic_capacity.csv')
wp_capacity['date'] = pd.to_datetime(wp_capacity['date'], format='%Y-%m-%d')
wp_capacity = wp_capacity.set_index('date')

# wp_capacity_s = (wp_capacity['capacity'].resample('Y').mean()).round(2)
# wp_capacity_s = (wp_capacity['capacity'].resample('Q').mean()).round(2)
# wp_capacity_s = (wp_capacity['capacity'].resample('W').mean()).round(2)
# wp_capacity_s = (wp_capacity['capacity'].resample('M').mean()).round(2)
wp_capacity_s = (wp_capacity['capacity']).round(2)

plt.figure(
    figsize = (16, 8),
    dpi = 100
)

ydata = wp_capacity_s
plt.plot(ydata, color='green')

plt.xticks(rotation=45)

plt.xlabel("date")
plt.ylabel("capacity")
plt.title("capacity")
# plt.legend(loc='lower right') # 标签位置

plt.savefig('capacity_D.jpg')

plt.show()
```
